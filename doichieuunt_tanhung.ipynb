#### Group các giao dịch PU theo tháng hạch toán.

```python
import pandas as pd
#read dataframe encode utf-8
df = pd.read_excel(r'.\input\sodathutheobienlai_tanhung.xlsx',header=0)
df['Ngày hạch toán'] = pd.to_datetime(df['Ngày hạch toán'])
df['Tháng'] = df['Ngày hạch toán'].dt.to_period('M')
df = df.groupby('Tháng').agg({'Số tiền hạch toán':'sum'}).reset_index()
df['Số tiền hạch toán'] = df['Số tiền hạch toán'].astype(float)
display(df)
```

```python
df_p9 = pd.read_excel(r'.\input\chungtuP9_tanhung2023.xlsx',header=0)
print(df_p9.columns)
#group số tiền theo tháng dựa vào ngày nộp thuế
df_p9['Ngày nộp thuế'] = pd.to_datetime(df_p9['Ngày nộp thuế'])
df_p9['Tháng'] = df_p9['Ngày nộp thuế'].dt.to_period('M')
df_p9 = df_p9.groupby('Tháng').agg({'Số tiền':'sum'}).reset_index()
df_p9['Số tiền'] = df_p9['Số tiền'].astype(float)
display(df_p9)
```
